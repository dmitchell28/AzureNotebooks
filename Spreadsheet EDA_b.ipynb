```python
#dataset.pivot_table(index = 'City', values = 'Usage Score', aggfunc = agglist).style.format(format_dict)
newdf = dataset.pivot_table(index = 'City', values = ['Usage Score', 'Hits'], aggfunc = agglist)
newdf['count']= dataset.pivot_table(index = 'City', values = 'Hits', aggfunc = 'count')
newdf
```

```python
newdf = dataset.pivot_table(index = 'Usage Score Bucket', values = ['Hits', 'Usage Score'])
newdf['count']= dataset.pivot_table(index = 'Usage Score Bucket', values = 'Hits', aggfunc = 'count')

format_dict = {'Hits':'{:.2f}', 'Usage Score':'{:.0f}'}
newdf.style.format(format_dict).set_caption('Hits BY BUCKET')
```
